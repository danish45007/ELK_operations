```python
# creating a es instance
endpoint = "http://localhost:9200/"
es = Elasticsearch(timeout=600,hosts=endpoint)
```

```python
df_new = df.to_dict("records")
```

```python
def generator(df):
    for i,line in enumerate(df):
        yield {
            '_index': 'netflix_data',
            '_type': '_doc',
            '_id': line.get("show_id",None),
            '_source': {
                "title": line.get("title",""),
                "director": line.get("director",""),
                "description": line.get("description",""),
                "cast": line.get("cast",""),
                "duration": line.get("duration",None)
            }
        }
#         raise StopIteration
```

```python
obj = generator(df_new)
```

```python
from pprint import pprint
pprint(next(obj))
```

```python
from elasticsearch import helpers
```

```python
# using bulk api to uplaod data onto the elasticsearch
try:
    res = helpers.bulk(es, generator(df_new))
    print("Uploaded!")
except Exception as e:
    print("Error:",e)
```

```python
# setting up custom mapping schema for the data
settings = {
   "settings":{
      "number_of_shards":1,
      "number_of_replicas":0
   },
   "mappings":{
      "properties":{
         "director":{
            "type":"text"
         },
         "duration":{
            "type":"text"
         }
      }
   }
}
```

```python
res = es.indices.create(index="netflix_data", ignore=[400,404], body=settings)
```
